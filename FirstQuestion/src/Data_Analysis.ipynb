```python
import pandas as pd #importing the required libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
data = pd.read_csv('CleanedDataSet.csv')

# Show the first few rows to verify the data
print(data.head())
```

```python
plt.figure(figsize=(8,6))
plt.hist(data['Age'], bins=8, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Strip any extra spaces from column names
data.columns = data.columns.str.strip()

# Now check the columns again
print(data.columns)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data['Height'], data['Weight'], color='coral')
plt.title('Height vs Weight')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (lbs)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Frailty', y='Grip strength', data=data, palette='Set2')
plt.title('Grip Strength by Frailty')
plt.xlabel('Frailty (0 = Not Frail, 1 = Frail)')
plt.ylabel('Grip Strength')
plt.show()
```

```python
plt.figure(figsize=(10,8))
correlation_matrix = data.corr()

# Create a heatmap with annotations
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(data, hue='Frailty', palette='coolwarm')
plt.show()
```
